```python
import pandas as pd
import numpy as np
import requests
from selenium import webdriver
#from bs4 import BeautifulSoup as bs
from selenium.webdriver import ActionChains
#from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
#from selenium.webdriver.support import expected_conditions as EC
import time
import pyperclip
from selenium.webdriver import ChromeOptions
#from selenium.webdriver.common.keys import Keys
import smtplib
import socks
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

#opening in incognito and removing testing info pop-up
opts=ChromeOptions()
opts.add_argument("--incognito")
opts.add_argument("--disable-infobars")
browser = webdriver.Chrome('chromedriver',options=opts)
browser.maximize_window()
browser.get("https://news.google.com/search?q=IOT%20bangladesh%20when%3A7d&hl=en-US&gl=US&ceid=US%3Aen")
#main_page = browser.current_window_handle
xps=[]
txts=[]
#making xpaths for clickable web elements for share button and copying text
for i in range(1,6):
    share_xpath='//*[@id="yDmH0d"]/c-wiz/div/div[2]/div[2]/div/main/c-wiz/div[1]/div[{}]/div/article/div[2]/menu/span[1]/div/span'.format(i)
    xps.append(share_xpath)
    txt_xpath='//*[@id="yDmH0d"]/c-wiz/div/div[2]/div[2]/div/main/c-wiz/div[1]/div[{}]/div/article/h3/a'.format(i)
    txts.append(txt_xpath)
button_xpath='//*[@id="yDmH0d"]/div[9]/div/div[2]/span/div[2]/div[2]/div[1]/button/span'

#close_xpath='//*[@id="yDmH0d"]/div[10]/div/div[2]/div[2]/div[3]/div/span/span'
art=0
j=0
try:
    links=[]
    texts=[]
    for x in xps:
    #putting the Action chains in the loop so it gets updated everytime
        acts=ActionChains(browser)
        browser.implicitly_wait(30)
        #clicking share button
        web_elem=browser.find_element_by_xpath(x)
    #     acts.move_to_element(web_elem)
        acts.click(web_elem)
        acts.perform()
    #     wait = WebDriverWait(browser, 60)
    #     button=wait.until(EC.presence_of_element_located((By.XPATH,button_xpath)))
        time.sleep(30)
        #clicking copy button
        button=browser.find_element_by_xpath(button_xpath)
        time.sleep(5)
    #     acts.move_to_element(button)
        acts.click(button)
        acts.perform()
    #     button.click()
        #txt=acts.send_keys(Keys.CONTROL,'v')
        time.sleep(5)
        #pasting the copied text
        links.append(pyperclip.paste())
        #getting the text of the html links
        time.sleep(5)
        for t in txts:
            texts.append(browser.find_element_by_xpath(t).get_attribute('text'))
    #     browser.switch_to_window(main_page)
    #     browser.refresh()
        time.sleep(5)
    #     close_butt=browser.find_element_by_xpath(close_xpath)
    #     acts.click(close_butt).perform()
        #refreshing the browser to find the next iterable xpaths in JS; this is a crucial step
        browser.refresh()
        j+=5
        #scrolling to see the clicked url properly
        browser.execute_script("window.scrollTo(0, {})".format(j))
        time.sleep(5)
        art+=1
        print("Article scrapped: {}".format(art))
    print("All 5 News are scrapped! Thanks to Rimon!")
    browser.quit()
except:
    print("Errors occured! please re-try or contact Rimon.")
    

#specifying addresses
#sender_gmail='IOTnewsscraper@gmail.com'
sender_email='IOT.Bangladesh@newsscraper.com'
#receiver_gmail=['','']
receiver_email=['']
subj="Weekly IOT News"

html="""\
<html>
  <head></head>
  <body>
    <p>Hi! I am <b>Keras Wazowski</b>, your news provider.
       Here is this week's sneak peek about IOT and Bangladesh:<br>
       <br>
    
       1. <a href="{0}" style="text-decoration:none">{1}</a>
       <br>
       <br>
       2. <a href="{2}" style="text-decoration:none">{3}</a>
       <br>
       <br>
       3. <a href="{4}" style="text-decoration:none">{5}</a>
       <br>
       <br>
       
       4. <a href="{6}" style="text-decoration:none">{7}</a>
       <br>
       <br>
       
       5. <a href="{8}" style="text-decoration:none">{9}</a>
       <br>
       <br>
       
       <strong>This is an auto-generated mail from Brain Birds Bangladesh. Please do not reply. Thanks!</strong><br>
       
    </p>
  </body>
</html>
""".format(links[0],texts[0],links[1],texts[1],links[2],texts[2],links[3],texts[3],links[4],texts[4])

try:
    mess=MIMEMultipart('alternative')
    #mess['From']='IOT Bangladesh'
    #To only takes string so joining strings from list
    #mess['To']=', '.join(receiver_gmail)
    mess['To']=', '.join(receiver_email)
    mess['Subject']=subj
    mess['cc']=''
#     mess.attach(MIMEText(body,'plain'))
    mess.attach(MIMEText(html,'html'))
    message=mess.as_string()
    #Starting the smtp connection
    #server=smtplib.SMTP('smtp.gmail.com',587)
    server=smtplib.SMTP('',25)
    server.ehlo()
#     server.starttls()
#     server.ehlo()
#     server.login(user=sender_gmail,password='')
    #server.set_debuglevel(1)
    #to_addrs takes list; no issue
    #server.sendmail(from_addr=sender_gmail,to_addrs=receiver_gmail,msg=message)
    server.sendmail(from_addr=sender_email,to_addrs=receiver_email,msg=message)
    server.quit()
    print("Mail is sent!")
except Exception as e:
    print("Failed to send mail! please re-try. Here is the error code: {}".format(e))
```
